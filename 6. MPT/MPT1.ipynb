```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 600
```

```python
# inputs
r1, r2, rf = 15, 10, 6
s1, s2, s12 = 500, 250, 100

# generte w's:
n = 1000
w1 = np.arange(1, n+1)/n
w2 = 1-w1

# returns and std:
rp = r1*w1 + r2*w2
sp = np.sqrt(w1**2 * s1 + w2**2 * s2 + 2*s12*w1*w2)

plt.scatter(sp, rp, marker='.', s=2)
plt.show()
```

```python
max_CAL = max((rp-rf)/sp)
s_cal = np.arange(0,n)*22/n
CAL_line = rf + max_CAL*s_cal

plt.scatter(sp, rp, marker='.', s=2)
plt.plot(s_cal, CAL_line, c='gray', linewidth=2)
plt.show()
```
